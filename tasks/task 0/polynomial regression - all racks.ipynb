# Fit the model using Polynomial Regression method

### Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Prepare the data

Upload gradient data

```python
data = pd.read_csv('files/transient_exp_data.csv')
print(data.shape)
print(data.head())
```

### Fill variables
- X refers to the independent values
- Y refers to the dependent values

```python
x_label = 'Time(s)'
y_labels = data.columns[1:]
```

```python
for y_label in y_labels:
    X = data[x_label].values
    Y = data[y_label].values
    
    #prepare the model
    coefficents = np.polyfit(X, Y, 3)
    p = np.poly1d(coefficents)
    
    # Evaluating the model Calculating Root Mean Squares Error
    rmse = 0
    n = len(X)
    for i in range(n):
        y_pred = p(X[i])
        rmse += (Y[i] - y_pred) ** 2

    rmse = np.sqrt(rmse/n)
    
    # Evaluating the model Calculating R2 Score
    # Mean X and Y
    mean_x = np.mean(X)
    mean_y = np.mean(Y)

    ss_tot = 0
    ss_res = 0
    for i in range(n):
        y_pred = p(X[i])
        ss_tot += (Y[i] - mean_y) ** 2
        ss_res += (Y[i] - y_pred) ** 2
    r2 = 1 - (ss_res/ss_tot)
    
    #plot
    fig = plt.figure(dpi=150)
    
    plt.plot(X, Y, 'o', label=y_label + ' data')
    plt.plot(X, p(X), 'r', label='Fitted line')

    plt.xlabel(x_label + "\n RMSE: " + str(rmse) + "\n R2 score: " + str(r2))
    plt.ylabel(y_label + " temperature in C")

    plt.title("model Fitted with polynomial regression")
    plt.suptitle(y_label)
    
    plt.legend()
    plt.show()

    
    
```
