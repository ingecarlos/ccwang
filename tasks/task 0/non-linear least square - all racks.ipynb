# Fit the model usen Least square

### Import libraries

```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
import pandas as pd
```

### Prepare the data

Upload gradient data

```python
data = pd.read_csv('files/transient_exp_data.csv')
print(data.shape)
print(data.head())
```

### Build the model
Prepare the function to be fitted: a 3th grade polynomial function

```python
def func(x, c0, c1, c2, c3):
    return c0 + c1*x + c2*x**2 + c3*x**3
```

### Fill variables
- X refers to the independent values
- Y refers to the dependent values

```python
x_label = 'Time(s)'
y_labels = data.columns[1:]
```

```python
for y_label in y_labels:
    X = data[x_label].values
    Y = data[y_label].values
    
    #fit the curve
    popt, pcov = curve_fit(func, X, Y)
    
    # Evaluate the model Calculating Root Mean Squares Error
    rmse = 0
    n = len(X)
    for i in range(n):
        y_pred = func(X[i], *popt)
        rmse += (Y[i] - y_pred) ** 2

    rmse = np.sqrt(rmse/n)
    
    # Evaluate the model Calculating R2 Score
    # Mean X and Y
    mean_x = np.mean(X)
    mean_y = np.mean(Y)

    ss_tot = 0
    ss_res = 0
    for i in range(n):
        y_pred = func(X[i], *popt)
        ss_tot += (Y[i] - mean_y) ** 2
        ss_res += (Y[i] - y_pred) ** 2
    r2 = 1 - (ss_res/ss_tot)
    
    #plot
    plt.plot(X, Y, 'o', label=y_label + ' data')
    plt.plot(X, func(X, *popt), 'r', label='Fitted line')

    plt.xlabel(x_label + "\n RMSE: " + str(rmse) + "\n R2 score: " + str(r2))
    plt.ylabel(y_label + " temperature in ◦C")

    plt.legend()
    plt.show()
```
