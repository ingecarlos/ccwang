# Fit the model usen Least square

### Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Prepare the data

Upload gradient data

```python
data = pd.read_csv('files/transient_exp_data.csv')
print(data.shape)
print(data.head())
```

### Fill variables
- X refers to the independent values
- Y refers to the dependent values

```python
x_label = 'Time(s)'
y_labels = data.columns[1:]
```

```python
for y_label in y_labels:
    X = data[x_label].values
    Y = data[y_label].values
    
    #prepare model
    A = np.vstack([X, np.ones(len(X))]).T
    m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

    #prepare regression
    regression = m*X + c
    
    #evaluate the model calculating root mean squares
    rmse = 0
    n = len(X)
    for i in range(n):
        y_pred = c + m * X[i]
        rmse += (Y[i] - y_pred) ** 2

    rmse = np.sqrt(rmse/n)
    
    #evaluate the model Calculating R2 Score
    mean_x = np.mean(X)
    mean_y = np.mean(Y)

    ss_tot = 0
    ss_res = 0
    for i in range(n):
        y_pred = c + m * X[i]
        ss_tot += (Y[i] - mean_y) ** 2
        ss_res += (Y[i] - y_pred) ** 2
    r2 = 1 - (ss_res/ss_tot)
    
    plt.plot(X, Y, 'o', label=y_label + ' data')
    plt.plot(X, regression, 'r', label='Fitted line')

    plt.xlabel(x_label + "\n RMSE: " + str(rmse) + "\n R2 score: " + str(r2))
    plt.ylabel(y_label + " temperature in ◦C")


    plt.title("model Fitted with least square")
    plt.suptitle(y_label)

    plt.legend()
    plt.show()
```
